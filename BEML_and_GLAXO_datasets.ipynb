```python
import pandas as pd
import numpy as np
```

```python
beml_df=pd.read_csv("BEML.csv")
glaxo_df=pd.read_csv("GLAXO.csv")
```

```python
beml_df.head()
```

```python
glaxo_df.head()
```

```python
beml_df=beml_df.loc[0:,"Date":"Close"]
glaxo_df=glaxo_df.loc[0:,"Date":"Close"]
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
beml_df["Date"]=pd.DatetimeIndex(beml_df["Date"])
glaxo_df["Date"]=pd.DatetimeIndex(glaxo_df["Date"])
```

```python
beml_df.dtypes
```

```python
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df["Date"]))
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df["Date"]))
```

```python
glaxo_df.dtypes
```

```python
glaxo_df.head()
```

```python
beml_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
plt.plot(glaxo_df.Close)
plt.xlabel("Time")
plt.ylabel("Closing Price")
```

```python
plt.plot(beml_df.Close)
plt.xlabel("Time")
plt.ylabel("Closing Price")
```

```python
beml_df["Gain"]=beml_df.Close.pct_change(periods=1)
glaxo_df["Gain"]=glaxo_df.Close.pct_change(periods=1)
```

```python
plt.plot(beml_df.Gain)
plt.xlabel("Time")
plt.ylabel("Gain")
plt.title("BEML")
```

```python
plt.plot(glaxo_df.Gain)
plt.xlabel("Time")
plt.ylabel("Gain")
plt.title("GLAXO")
```

```python
sn.distplot(glaxo_df.Gain,label="GLAXO")
sn.distplot(beml_df.Gain,label="BEML")
plt.legend()
plt.show()
```

```python
beml_mean=round(beml_df.Gain.mean(),4)
beml_std=round(beml_df.Gain.std(),4)
glaxo_mean=round(glaxo_df.Gain.mean(),4)
glaxo_std=round(glaxo_df.Gain.std(),4)
```

```python
beml_std
```

```python
glaxo_std
```

we can that beml has high variation than glaxo

# ***which has more probability that returns 2% or more daily return***


```python
import scipy.stats as st
1-st.norm.cdf(0.02,beml_mean,beml_std)
```

```python
1-st.norm.cdf(0.02,glaxo_mean,glaxo_std)
```

BEML has more probablity

# ***which has more probability that returns less than 2% daily return***


```python
st.norm.cdf(-0.02,beml_mean,beml_std)
```

```python
st.norm.cdf(-0.02,glaxo_mean,glaxo_std)
```

GLAXO has more probablity
